**Defing the coloums to plot**

```python
Tem = df['Temperature']
Resistance = df['Resistance']
```

**Plotting the graph and finding the slope of the gaph**

```python
scatter(Tem,Resistance,color='red',label='Data')
coefficients = polyfit(Tem,Resistance, 1) #Command for finding the slope
slope = coefficients[0]
plot(Tem, polyval(coefficients, Tem), color='black', label=f'Trendline (Slope: {slope})')
xlabel('Temeprature ($^oC$)')
ylabel('Resistance ($\Omega$)')
title('Fermi Energy')
legend()
savefig('Fermi_graph')
show()
```
