**Importing the required libraries**

```python
from numpy import *
from matplotlib.pyplot import *
from math import factorial
```

**Defining the eigen functions for ground state and also defining the ladder operator**

```python
def func(x):
    return (1/sqrt(pi))*exp(-(1/2)*x**2)

def ladder(func,x):
    x0 = 1
    y = (x*func - gradient(func,x))
    return y
```

**Writting the logics to plot first five eigenfunction and the corresponding probability density**

```python
x = linspace(-6,6,1000)  #Defining the x-values
n = 6 #This determines the number of eigenfunctions we are plotting
funcn = func(x) #Defining the ground-state eigenfunction
#plotting the ground-state eigenfuntion
fig , axs = subplots(1, 2, figsize=(14, 6)) 
axs[0].plot(x,funcn,label=f'n = 1')
axs[0].set_title('Eigen function',fontsize=15)
axs[0].set_xlabel('X values',fontsize=15)
f = funcn**2 #Getting the probability densities
axs[1].plot(x,f,label=f'n = 1')
axs[1].set_title('Probrability Distribution',fontsize=15)
axs[1].set_xlabel('X values',fontsize=15)
tight_layout()
legend()
savefig('n = 1') #Saving the image for the ground state eigenfunction
#Starting the loop for plotting the higer order eigenfunction using the ladder operator
i = 2
while i <= n:
    funcn = ladder(funcn,x,i)/sqrt(sqrt(factorial(i))) #Calculating the higher order eigen functions using the ladder function defined
    #Plotting the higher order wavefunction and corresponding probability densities
    fig , axs = subplots(1, 2, figsize=(14, 6))
    axs[0].plot(x,funcn,label=f'n = {i}')
    axs[0].set_title('Eigen function',fontsize=15)
    axs[0].set_xlabel('X values',fontsize=15)
    f = funcn**2 #Getting the probability densities
    axs[1].plot(x,f,label=f'n = {i}')
    axs[1].set_title('Probrability Distribution',fontsize=15)
    axs[1].set_xlabel('X values',fontsize=15)
    tight_layout()
    legend()
    savefig(f'n = {i}') #Saving the plots
    i+=1
show()
```
