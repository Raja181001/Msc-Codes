```python
import matplotlib.pyplot as plt
import numpy as np

def optical(m1, m2, k, a):
    return np.sqrt((1 / (m1 * m2)) * (m1 + m2 + np.sqrt(m1**2 + m2**2 + 2 * m1 * m2 * np.cos(k * a))))

def acoustical(m1, m2, k, a):
    return np.sqrt((1 / (m1 * m2)) * (m1 + m2 - np.sqrt(m1**2 + m2**2 + 2 * m1 * m2 * np.cos(k * a))))

# Parameters
k = np.pi / a  # Assuming 'a' is defined
m1_over_m2_values = np.linspace(0.00001, 100, 100)  # Varying the ratio of m1/m2 from 1 to 10

# Set a constant value for 'a'
a = 1.0

# Calculate frequency differences for each 'm1/m2'
frequency_differences = []

for m_ratio in m1_over_m2_values:
    optical_frequencies = optical(m_ratio,0.000001, k, a)
    acoustical_frequencies = acoustical(m_ratio, 0.000001, k, a)
    frequency_difference = optical_frequencies - acoustical_frequencies
    frequency_differences.append(frequency_difference)

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(m1_over_m2_values, frequency_differences)
plt.title('Difference between Optical and Acoustical Frequencies at k = pi/a')
plt.xlabel('Mass Ratio (m1/m2)')
plt.ylabel('Forbidden gap')
plt.show()
```
