**Importing the required libraries**

```python
from numpy import *
from matplotlib.pyplot import *
```

**Initializing the Amplitude, wave vector, angular frequency, phase for the Electric filed**

```python
#Defining amplitude for the Electric filed
E_0 = 1 #Amplitude of Electric filed 1
E_01 = 2 #Amplitude of Electric filed 2
#Defining the wave vector
k1 = 2 #Wave vector of Electric filed 1
k2 = 2 #wave vector of Electric filed 2
#Defining the angular frequency
omega1 = 2 #Angular frequency of Electric filed 1
omega2 = 2 #Angular frequency of Electric filed 2
#Defining the phase 
phase1 = pi/2 #Phase of Electric filed 1
phase2 = 0 #Phase of Electric filed 2
```

**Defining the Electric as a function of time at a particular position in space**

```python
t = linspace(0,10,1000)
x = 1
E_1 = E_0 * sin(k1*x - omega1*t + phase1)
E_2 = E_01* sin(k2*x - omega2*t + phase2)
```

**Plotting the Electric filed in 2D and in 3D graph**

```python
fig = figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot(zeros_like(t),E_1,t,label='E_1')
ax.plot(E_2,zeros_like(t),t,label='E_2')
ax.plot([min(E_1), max(E_1)], [0, 0], [0, 0], c='black', linestyle='--' )
ax.plot([0, 0], [min(E_2), max(E_2)], [0, 0], c='black', linestyle='--')
ax.plot([0, 0], [0, 0], [min(t), max(t)], c='black', linestyle='--')
ax.set_xlabel('X-axis', labelpad=0)  
ax.set_ylabel('Y-axis', labelpad=0)  
ax.set_zlabel('Z-axis', labelpad=-30)  
ax.set_xticks([]) 
ax.set_yticks([]) 
ax.set_zticks([])
ax.view_init(elev=60, azim=30)
ax.grid(False)
legend()
#savefig(f'/run/media/rajapandi/New Volume/Entertainment/St. Xaviers college/MSc/Sem 2/Practicals/Electrodynamics/Practical 3/initial3d_{phase1}_{phase2}_{E_01}_{E_0}.png')
show()
plot(t,E_1)
xlabel('Direction of propogation')
ylabel('Direction of vibration')
plot([0, 0],[min(E_1), max(E_1)],  c='black', linestyle='--' )
plot( [min(t), max(t)],[0, 0], c='black', linestyle='--')
xticks([]) 
yticks([])
#savefig(f'/run/media/rajapandi/New Volume/Entertainment/St. Xaviers college/MSc/Sem 2/Practicals/Electrodynamics/Practical 3/intial_E1_2d_{phase1}_{phase2}_{E_01}_{E_0}.png')
show()
plot(t,E_2)
plot([0, 0],[min(E_2), max(E_2)],  c='black', linestyle='--')
plot([min(t), max(t)],[0,0], c='black', linestyle='--')
xlabel('Direction of Propogation')
ylabel("Direction of Vibration")
xticks([]) 
yticks([])
#savefig(f'/run/media/rajapandi/New Volume/Entertainment/St. Xaviers college/MSc/Sem 2/Practicals/Electrodynamics/Practical 3/intial_E2_2d_{phase1}_{phase2}_{E_01}_{E_0}.png')
show()
```

**Plotting the Polarized Electric filed**

```python
fig = figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot(E_2, E_1, t, label='Polarised E', color='red')
if max(E_1) <= max(E_2):
    xlim(-max(E_2)-0.2,max(E_2)+0.2)
    ylim(-max(E_2)-0.2,max(E_2)+0.2)
else:
    xlim(-max(E_1)-0.2,max(E_1)+0.2)
    ylim(-max(E_2)-0.2,max(E_2)+0.2)
ax.plot([min(E_1), max(E_1)], [0, 0], [0, 0], c='black', linestyle='--' )
ax.plot([0, 0], [min(E_2), max(E_2)], [0, 0], c='black', linestyle='--')
ax.plot([0, 0], [0, 0], [min(t), max(t)], c='black', linestyle='--')
ax.set_xlabel('X-axis', labelpad=0)  
ax.set_ylabel('Y-axis', labelpad=0)  
ax.set_zlabel('Z-axis', labelpad=-30)  
ax.set_xticks([]) 
ax.set_yticks([]) 
ax.set_zticks([])
ax.view_init(elev=90, azim=90)
ax.grid(False)
legend()
#savefig('/run/media/rajapandi/New Volume/Entertainment/St. Xaviers college/MSc/Sem 2/Practicals/Electrodynamics/Practical 3/Linear_45_o.png')
show()
```
