**Importing the required Libraries**

```python
from numpy import *
from matplotlib.pyplot import *
```

**Defining the functions for calculating the acceleration velocity and position**

```python
def acc(r):
    epsilon = 0.0103
    sigma = 3.4
    m = 39.948
    f = - 4 * epsilon * (((6) /(r**7)) * (sigma)**6 - ((12)/(r**13))* (sigma)**12)
    a = f/m
    return a

def velocity(v,a_1,a_2):
    dt = 0.001
    v1 = v + (1/2) * (a_1 + a_2) * dt
    return v1

def position(x,v,a):
    dt = 0.001
    x1 = x + v * dt + (1/2)*a * dt**2
    return x1
```

**Code for fiding the acceleration, velocity and position of the particle**

```python
#Initializing thei initial position and initial velocity
x1 = [1] #position for particle 1
x2 = [6] #position for particle 2
x3 = [12] #position for particle 3
v1 = [0.01] #velocity for particle 1
v2 = [0] #velocity for particle 2
v3 = [0.001] #velocity for particle 3
n = [0] #number of interactions
#starting a loop for finding the acceleration, velocity and position for every itteration
i = 1 
while i < 1000000:
    x1.append(position(x1[i-1],v1[i-1],acc(x1[i-1] - x2[i-1])+acc(x1[i-1] - x3[i-1])))
    x2.append(position(x2[i-1],v2[i-1],acc(x2[i-1] - x1[i-1])+acc(x2[i-1] - x3[i-1])))
    x3.append(position(x3[i-1],v3[i-1],acc(x3[i-1] - x1[i-1])+acc(x3[i-1] - x2[i-1])))
    v1.append(velocity(v1[i-1],acc(x1[i-1] - x2[i-1])+acc(x1[i-1] - x3[i-1])
                       ,acc(x1[i] - x2[i])+acc(x1[i] - x3[i])))
    v2.append(velocity(v2[i-1],acc(x2[i-1] - x1[i-1])+acc(x2[i-1] - x3[i-1])
                       ,acc(x2[i] - x1[i])+acc(x2[i] - x3[i])))
    v3.append(velocity(v3[i-1],acc(x3[i-1] - x1[i-1])+acc(x3[i-1] - x2[i-1])
                       ,acc(x3[i] - x1[i])+acc(x3[i] - x2[i])))
    n.append(i)
    i+=1
```

**Code for plotting the graphs**

```python
plot(n,x1,'--',color='blue',label='Trajectory of particle 1')
plot(n,x2,'--',color='red',label='Trajectory of particle 2')
plot(n,x3,'--',color='green',label='Trajectory of particle 3')
scatter(n[0],x1[0],color='blue')
scatter(n[0],x2[0],color='red')
scatter(n[0],x3[0],color='green')
scatter(n[-1],x1[-1],color='blue')
scatter(n[-1],x2[-1],color='red')
scatter(n[-1],x3[-1],color='green')
xlabel('number of steps',fontsize=15)
ylabel('Position',fontsize=15)
legend()
savefig('/run/media/rajapandi/New Volume/Entertainment/St. Xaviers college/MSc/Sem 2/Practicals/Statistical Mechanics/Practical 2/output.png')
show()
```
