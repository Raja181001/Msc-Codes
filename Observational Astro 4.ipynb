**This python codde is to remove all Divide by zero error and plot the velocity Vs Radiometer Temperature graph**

```python
from numpy import *
from pandas import *
from matplotlib.pyplot import *
```

```python
file_path = '/run/media/rajapandi/New Volume/Entertainment/St. Xaviers college/MSc/Sem 2/Practicals/Observational Astronomy/Practical 4/Readings/radio_milkyway.csv'
df = read_csv(file_path)
```

```python

df['t_s'] = df['t_s'].replace('#VALUE!', np.nan)

# Identify rows with NaN values in the 'y' column
nan_mask_y = df['t_s'].isna()

# Get the indices of rows with NaN values in the 'y' column
removed_indices_y = df.index[nan_mask_y].tolist()

# Remove NaN values from the DataFrame
df_cleaned_y = df.dropna(subset=['t_s'])

# Now, df_cleaned_y is your DataFrame without NaN values in the 'y' column,
# and removed_indices_y contains the indices of removed rows in the 'y' column.

# If you want to remove rows with '#DIV/0!' from the entire DataFrame, you can use:
df_cleaned = df.replace('#VALUE!', np.nan).dropna()
```

```python
x_axis = df_cleaned['t_s']
y_axis = df_cleaned['velocity']

# Zip the lists and sort them based on the values in list1
sorted_lists = sorted(zip(y_axis, x_axis), key=lambda x: x[0])

# Unpack the sorted pairs into separate lists
sorted_list1, sorted_list2 = zip(*sorted_lists)
```

```python
# Convert list2 to float
y_axis = [float(value) for value in sorted_list2]

# Plotting
plot(sorted_list1,y_axis)
xlabel('Velocity')
ylabel('Radiometer Temperature')
title('Velocity Vs Radiometer Temperature')
#savefig('Astro_4')
show()
```
